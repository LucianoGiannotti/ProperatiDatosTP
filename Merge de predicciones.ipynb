```python
import pandas as pd
```

```python
pred_xgboost = pd.read_csv('Predicciones/XgBoost.csv')
```

```python
pred_adaboost = pd.read_csv('Predicciones/Ada_Boost.csv')
pred_adaboost.rename(columns = {'id': 'id','price_aprox_usd':'price_usd'},inplace=True)
```

```python
pred_xgboost
```

```python
pred_adaboost
```

```python
pred_adaboost.id.values
```

```python
ids_ada = pred_adaboost.id.values
ids_xgb = pred_xgboost.id.values

ids_faltantes_en_ada = list()
ids_faltantes_en_ada.extend(x for x in ids_xgb if x not in ids_ada)
```

```python
ids_faltantes_en_ada
```

```python
pred_merge = pred_adaboost.append(pred_xgboost[pred_xgboost['id'].isin(ids_faltantes_en_ada)])
```

```python
pred_merge.sort_values('id',axis=0,inplace=True)
```

```python
pred_merge
```

```python
pred_merge.to_csv('Predicciones/Ada_Boost+XGBOOST.csv', index = False)
```
