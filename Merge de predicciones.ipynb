### Verifico que a todos les falten las misma cantidad y numeros de ids

```python
ids_extra_tree = pred_extra_tree.id.values
ids_xgb = pred_xgboost2.id.values
ids_faltantes_en_extra_tree = list()
ids_faltantes_en_extra_tree.extend(x for x in ids_xgb if x not in ids_extra_tree)


ids_extra_tree2 = pred_extra_tree2.id.values
ids_faltantes_en_extra_tree2 = list()
ids_faltantes_en_extra_tree2.extend(x for x in ids_xgb if x not in ids_extra_tree2)


ids_extra_tree3 = pred_extra_tree3.id.values
ids_faltantes_en_extra_tree3 = list()
ids_faltantes_en_extra_tree3.extend(x for x in ids_xgb if x not in ids_extra_tree3)


ids_extra_tree4 = pred_extra_tree4.id.values
ids_faltantes_en_extra_tree4 = list()
ids_faltantes_en_extra_tree4.extend(x for x in ids_xgb if x not in ids_extra_tree4)


ids_extra_tree5 = pred_extra_tree5.id.values
ids_faltantes_en_extra_tree5 = list()
ids_faltantes_en_extra_tree5.extend(x for x in ids_xgb if x not in ids_extra_tree5)


ids_ada = pred_adaboost.id.values
ids_faltantes_en_ada = list()
ids_faltantes_en_ada.extend(x for x in ids_xgb if x not in ids_ada)

ids_random_forest = pred_random_forest.id.values
ids_faltantes_en_random_forest = list()
ids_faltantes_en_random_forest.extend(x for x in ids_xgb if x not in ids_random_forest)


ids_decission_tree = pred_decission_tree.id.values
ids_faltantes_en_decission_tree = list()
ids_faltantes_en_decission_tree.extend(x for x in ids_xgb if x not in ids_decission_tree)


ids_decission_tree2 = pred_decission_tree2.id.values
ids_faltantes_en_decission_tree2 = list()
ids_faltantes_en_decission_tree2.extend(x for x in ids_xgb if x not in ids_decission_tree2)
```

```python
print ((ids_faltantes_en_ada == ids_faltantes_en_decission_tree)
       and (ids_faltantes_en_ada == ids_faltantes_en_decission_tree2)
       and (ids_faltantes_en_ada == ids_faltantes_en_extra_tree) 
       and (ids_faltantes_en_ada == ids_faltantes_en_extra_tree2) 
       and (ids_faltantes_en_ada == ids_faltantes_en_extra_tree3) 
       and (ids_faltantes_en_ada == ids_faltantes_en_extra_tree4) 
       and (ids_faltantes_en_ada == ids_faltantes_en_extra_tree5) 
       and (ids_faltantes_en_ada == ids_faltantes_en_random_forest) )
```

```python
ids_faltantes = ids_faltantes_en_ada
```

```python
pred_promedio = pred_adaboost.copy()
pred_promedio.rename(columns={"price_usd":"price1"},inplace=True)
pred_promedio["price2"] = pred_random_forest['price_usd']
pred_promedio["price3"] = pred_decission_tree['price_usd']
pred_promedio["price4"] = pred_decission_tree2['price_usd']
pred_promedio["price5"] = pred_extra_tree['price_usd']
pred_promedio["price6"] = pred_extra_tree2['price_usd']
pred_promedio["price7"] = pred_extra_tree3['price_usd']
pred_promedio["price8"] = pred_extra_tree4['price_usd']
pred_promedio["price9"] = pred_extra_tree5['price_usd']

pred_promedio
```

```python
pred_promedio['price_usd'] = pred_promedio["price1"] + pred_promedio["price2"] + pred_promedio["price3"] + \
                            pred_promedio["price4"] + pred_promedio["price5"] + pred_promedio["price6"] + \
                            pred_promedio["price7"] + pred_promedio["price8"] + pred_promedio["price9"]
        
pred_promedio.drop(["price1","price2","price3","price4","price5","price6","price7","price8","price9"],1,inplace=True)
pred_promedio
```
